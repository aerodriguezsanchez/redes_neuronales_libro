````python
# Cap. 2 - Programa 1
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.metrics import mse
````

````python
def sig(x):
  return 1 / (1 + np.exp(-x))
````

````python
def prop_ad(X):
  a_0 = np.hstack((np.ones((X.shape[0], 1)), X))
  a_1 = np.hstack((np.ones((X.shape[0], 1)),
  sig(np.dot(a_0, W_1))))
  a_L = np.dot(a_1, W_2)
  return a_0, a_1, a_L
````

````python
def retroprop(a_0, a_1, a_L):
  d_L = a_L - y
  d_1 =a_1[:,1:]*(1-a_1[:,1:])*np.dot(d_L, W_2.T[:, 1:])
  dc_dW_1 = a_0[:,:,np.newaxis] * d_1[:,np.newaxis,:]
  dc_dW_2 = a_1[:,:,np.newaxis] * d_L[:,np.newaxis,:]
  grad_C_W_1 = np.average(dc_dW_1, axis=0)
  grad_C_W_2 = np.average(dc_dW_2, axis=0)
  return grad_C_W_1, grad_C_W_2
````

````python
# Hiperparametros
tf.random.set_seed(1)
alfa = 0.01
neuronas_oc = 10
epocas = 1000
vec_epocas = np.zeros(epocas)
vec_error = np.zeros(epocas)
````

````python
# Datos de entrenamiento
N = 1000
X = tf.linspace([-1.], [10.], N)
y = 2 * X + 2
y_ruido = tf.random.normal((N, 1), 0.0, 2.0)
y = y + y_ruido
````

````python
# Descenso de gradiente
W_1 = 2*np.random.random((X.shape[1]+1, neuronas_oc))-1
W_2 = 2*np.random.random((neuronas_oc+1, y.shape[1]))-1

for i in range(epocas):
  # Propagacion:
  a_X, a_1, a_L = prop_ad(X)
  # Retropropagacion:
  grad_C_W_1, grad_C_W_2 = retroprop(a_X, a_1, a_L)
  # Actualizacion de pesos:
  W_1 += - alfa * grad_C_W_1
  W_2 += - alfa * grad_C_W_2
  # Vectores de error y epocas:
  error = mse(np.squeeze(y), np.squeeze(a_L))
  vec_epocas[i] = i
  vec_error[i] = error.numpy()
  print("Error: ", error.numpy())

print("Fin del entrenamiento")
print("Error final: ", error.numpy())
````

````python
plt.plot(vec_epocas, vec_error, 'ko-')
plt.xlabel("´Epocas")
plt.ylabel("$MSE$")
plt.show()
````

````python
plt.plot(X, y, 'o', label='Datos')
plt.plot(X, a_L, 'k-', label='Red neuronal')
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
````

````python
import pandas as pd
from tensorflow.keras.metrics import mse
from scipy import stats

path = 'https://github.com/aerodriguezsanchez/redes_neuronales_libro/raw/main/SOCR-HeightWeight.csv'
datos = pd.read_csv(path, sep=',')

x_prueba = datos.iloc[-1000:, 1].to_numpy()
y_prueba = datos.iloc[-1000:, 2].to_numpy()

x_A = datos.iloc[:10, 1].to_numpy()
y_A = datos.iloc[:10, 2].to_numpy()
x_B = datos.iloc[:1000, 1].to_numpy()
y_B = datos.iloc[:1000, 2].to_numpy()

res_A = stats.linregress(x_A, y_A)
res_B = stats.linregress(x_B, y_B)

y_A_pred = res_A.slope * x_prueba + res_A.intercept
y_B_pred = res_B.slope * x_prueba + res_B.intercept

MSE_A = mse(y_prueba, y_A_pred)
MSE_B = mse(y_prueba, y_B_pred)

print("MSE_A = ", MSE_A.numpy())
print("MSE_B = ", MSE_B.numpy())
print("RMSE_A = ", MSE_A.numpy() ** 0.5)
print("RMSE_B = ", MSE_B.numpy() ** 0.5)
````

````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Definicion y preparacion de los datos
path = "https://github.com/aerodriguezsanchez/redes_neuronales_libro/raw/main/slump_test.data"

datos = pd.read_csv(path, sep=',')
datos = datos.drop(['No'], axis=1)

modelo_1_x = datos.iloc[:, 0].to_numpy()
modelo_1_x = modelo_1_x[:, np.newaxis]
modelo_2_x = datos.iloc[:, [0, 3, 5]].to_numpy()
modelo_3_x = datos.iloc[:, [0, 1, 2, 3, 5]].to_numpy()

y = datos.iloc[:, 9].to_numpy()
y = y[:, np.newaxis]

# Creacion, compilacion y evaluacion de modelos

# Modelo 1 - k=1
k_1 = 3
n_1 = modelo_1_x.shape[0]

modelo_1_y = LinearRegression()
modelo_1_y.fit(modelo_1_x, y)

y_pred_1 = modelo_1_y.predict(modelo_1_x)

R2_1 = r2_score(y_pred_1, y)
R2_1_aj = 1 - (1 - R2_1) * (n_1 - 1) / (n_1 - k_1 - 1)
print("Result. Modelo 1 |> Rˆ2 = %.2f; Rˆ2_aj = %.2f" % (R2_1, R2_1_aj))

# Modelo 2 - k=3
k_2 = 3
n_2 = modelo_2_x.shape[0]

modelo_2_y = LinearRegression()
modelo_2_y.fit(modelo_2_x, y)

y_pred_2 = modelo_2_y.predict(modelo_2_x)

R2_2 = r2_score(y_pred_2, y)
R2_2_aj = 1 - (1 - R2_2) * (n_2 - 1) / (n_2 - k_2 - 1)
print("Result. Modelo 2 |> Rˆ2 = %.2f; Rˆ2_aj = %.2f" % (R2_2, R2_2_aj))

# Modelo 3 - k=5
k_3 = 5
n_3 = modelo_3_x.shape[0]

modelo_3_y = LinearRegression()
modelo_3_y.fit(modelo_3_x, y)

y_pred_3 = modelo_3_y.predict(modelo_3_x)
R2_3 = r2_score(y_pred_3, y)
R2_3_aj = 1 - (1 - R2_3) * (n_3 - 1) / (n_3 - k_3 - 1)

print("Result. Modelo 3 |> Rˆ2 = %.2f; Rˆ2_aj = %.2f" % (R2_3, R2_3_aj))
````

````python
print(f'Forma del tensor X: {X.shape}.')
print(f'Forma del tensor X entrenamiento: {X_train.shape}.')
print(f'Forma del tensor X validacion: {X_valid.shape}.')
print(f'Forma del tensor X prueba: {X_test.shape}.')
````

````python
X_train = X_train / 255.
X_test = X_test / 255.
````

````python
# Convertir etiquetas en one-hot
temp = []
for i in range(len(y_train)):
  temp.append(to_categorical(y_train[i], num_classes=10))
y_train = np.array(temp)

temp = []
for i in range(len(y_valid)):
  temp.append(to_categorical(y_valid[i], num_classes=10))
y_valid = np.array(temp)

temp = []
for i in range(len(y_test)):
  temp.append(to_categorical(y_test[i], num_classes=10))
y_test = np.array(temp)
````

````python
modelo.compile(loss='categorical_crossentropy',
               optimizer='SGD',
               metrics='acc')
````

````python
hist = modelo.fit(X_train,
                  y_train,
                  epochs=10,
                  validation_data=(X_valid, y_valid)
                  )
````

````python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow.keras.datasets.cifar10 as cifar10
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
````

````python
(X, y), (X_test, y_test) = cifar10.load_data()
(X_train, X_valid) = (X[:40000], X[40000:])
(y_train, y_valid) = (y[:40000], y[40000:])

X_train = X_train / 255.
X_valid = X_valid / 255.
X_test = X_test / 255.
````

````python
# Convertir etiquetas en one-hot
temp = []

for i in range(len(y_train)):
    temp.append(to_categorical(y_train[i], num_classes=10))
y_train = np.squeeze(np.array(temp))

temp = []
for i in range(len(y_valid)):
    temp.append(to_categorical(y_valid[i], num_classes=10))
y_valid = np.squeeze(np.array(temp))

temp = []
for i in range(len(y_test)):
    temp.append(to_categorical(y_test[i], num_classes=10))
y_test = np.squeeze(np.array(temp))
````

````python
CifarTinyVGG.compile(loss='categorical_crossentropy',
                     optimizer='SGD',
                     metrics='acc')
````

````python
perdida_TinyVGG = CifarTinyVGG.fit(X_train,
                                   y_train,
                                   epochs=1, # <-- Modificar epocas
                                   validation_data=(X_valid, y_valid)
                                   )
````

````python
test_error_TinyVGG = CifarTinyVGG.evaluate(x=X_test, y=y_test)

print("El accuracy es: %.4f" % test_error_TinyVGG[1])
````
